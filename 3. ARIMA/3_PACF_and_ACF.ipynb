```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Partial Autocorrelation

```python
np.random.seed(1)
x0 = np.random.randn(1000)
```

```python
plt.plot(x0);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_pacf(x0, ax=ax);
```

```python
# Creating AR(1) process to test PACF

#np.random.seed(1)
x1 = [0]
for i in range(1000):

  # Next value is 0.5 of previous value plus Gaussian noise with standard
  # deviation of 0.1
  x = 0.5 * x1[-1] + 0.1 * np.random.randn()
  x1.append(x)
x1=np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig, ax = plt.subplots(figsize= (10, 5))
plot_pacf(x1, ax=ax);
```

```python
x1 = [0]
for i in range(1000):

  # Next value is 0.5 of previous value plus Gaussian noise with standard
  # deviation of 0.1
  x = -0.5 * x1[-1] + 0.1 * np.random.randn()
  x1.append(x)
x1=np.array(x1)
```

```python
fig, ax = plt.subplots(figsize= (10, 5))
plot_pacf(x1, ax=ax);
```

```python
# AR(2) process
# Depends on two past values

x2 = [0, 0]
#np.random.seed(1)

for i in range(1000):
  x = 0.5 * x2[-1] - 0.3 * x2[-2] + 0.1 * np.random.randn()
  x2.append(x)
x2=np.array(x2)
```

```python
plt.plot(x2);
```

```python
fig, ax = plt.subplots(figsize= (10, 5))
plot_pacf(x2, ax=ax);
```

```python
x2 = [0, 0]
#np.random.seed(1)

for i in range(1000):
  x = -0.5 * x2[-1] - 0.3 * x2[-2] + 0.1 * np.random.randn()
  x2.append(x)
x2=np.array(x2)
```

```python
fig, ax = plt.subplots(figsize= (10, 5))
plot_pacf(x2, ax=ax);
```

```python
# AR(5) process
# next value of x depends on 3 lagged values

x5 = [0, 0, 0, 0, 0]
#np.random.seed(1)

for i in range(1000):
  x = 0.5 * x5[-1] - 0.3 * x5[-2] - 0.6 * x5[-5] + 0.1 * np.random.randn()
  x5.append(x)
x5=np.array(x5)
```

```python
plt.plot(x5);
```

```python
fig, ax = plt.subplots(figsize= (10, 5))
plot_pacf(x5, ax=ax);
```

### Autocorrelation

```python
# iid noise

fig, ax = plt.subplots(figsize= (10, 5))
plot_acf(np.random.randn(1000), ax=ax);
```

```python
# Generate Moving Average MA(1)
# Value is generated by adding current error + coef * prev error

errors = 0.1 * np.random.randn(1000)
ma1 = []

for i in range(1000):
  if i >= 1:
    x = 0.5 * errors[i-1] + errors[i]
  else:
    x = errors[i]
  ma1.append(x)
ma1 = np.array(ma1)
```

```python
plt.plot(ma1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(ma1, ax=ax);
```

```python
# Generate MA(2) process
# Value depends on three errors: current, lag one, and lag two

errors = 0.1 * np.random.randn(1000)
ma2 = []

for i in range(1000):
  x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + errors[i]
  ma2.append(x)
ma2 = np.array(ma2)
```

```python
plt.plot(ma2);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(ma2, ax=ax);
```

```python
# Generate MA(3) process
# Value depends on three errors: current, lag one, lag two, and lag three

errors = 0.1 * np.random.randn(1000)
ma3 = []

for i in range(1000):
  x = 0.5 * errors[i-1] - 0.3 * errors[i-2] - 0.7 * errors[i-3] + errors[i]
  ma3.append(x)
ma3 = np.array(ma3)
```

```python
plt.plot(ma3);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(ma3, ax=ax);
```

```python
# Generate MA(6) process
# Value depends on three errors: current, lag one to six

errors = 0.1 * np.random.randn(1000)
ma6 = []

for i in range(1000):
  x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + 0.7 * errors[i-3] +\
      0.2 * errors[i-4] - 0.8 * errors[i-5] - 0.9 * errors[i-6] + errors[i]
  ma6.append(x)
ma6 = np.array(ma6)
```

```python
plt.plot(ma6);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(ma6, ax=ax);
```
