```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Number of time steps

T=1000

# Initial price

P0=10

# Drift

mu=0.001
```

```python
# last_log_price

last_p = np.log(P0)

log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
  # Sample a log return
  r = 0.01 * np.random.randn()
  
  # Compute new log price
  p = last_p + mu + r

  # Store the return and price
  log_returns[t] = r+mu
  prices[t] = np.exp(p)
  
  # Assign last_p
  last_p = p
```

```python
plt.figure(figsize=(20, 8))
plt.plot(prices)
plt.title("Simulated Prices");
```
